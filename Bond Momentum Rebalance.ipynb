```python
import bt
import pandas as pd
%pylab inline
```

```python
def long_only_ew(tickers, start='2010-01-01', name='benchmark'):
    s = bt.Strategy(name, [bt.algos.RunOnce(),
                           bt.algos.SelectAll(),
                           bt.algos.WeighEqually(),
                           bt.algos.Rebalance()])
    data = bt.get(tickers, start=start)
    return bt.Backtest(s, data)

```

```python
class OrderedWeights(bt.Algo):
    def __init__(self, weights):
        self.target_weights = weights

    def __call__(self, target):
        target.temp['weights'] = dict(zip(target.temp['selected'], self.target_weights))
        return True
```

```python
class ApplyLeverage(bt.Algo):
    
    def __init__(self,leverage):
        super(ApplyLeverage,self).__init__()
        self.leverage = leverage
        
    def __call__(self,target):
        target.temp['weights'] = target.temp['weights']*self.leverage
        return True
```

```python
def meanVar(tickers, mons, name = 'meanVar', x=-1.0,y=1.0, start = '1980-01-01'):
    data = bt.get(tickers, start = start)
    s = bt.Strategy(name, [bt.algos.RunMonthly(),
                          bt.algos.SelectAll(),
                          #bt.algos.SelectMomentum(n=6),
                          bt.algos.WeighMeanVar(lookback = pd.DateOffset(months=mons), bounds=(x,y)),
                          ApplyLeverage(2.0),
                          bt.algos.Rebalance()])
    
    return bt.Backtest(s,data)

def nThree(tickers, name='nThree', start='2000-01-01', w=.5):
    data = bt.get(tickers, start = start)
    s = bt.Strategy(name, [bt.algos.RunMonthly(),
                          bt.algos.SelectAll(),
                          bt.algos.SelectMomentum(n=3),
                          OrderedWeights([w,w,w]),
                          bt.algos.Rebalance()])
    
    return bt.Backtest(s,data)

def nSix(tickers, name='nSix', start = '2000-01-01', w=.2 ):
    data = bt.get(tickers, start = start)
    s = bt.Strategy(name, [bt.algos.RunMonthly(),
                          bt.algos.SelectAll(),
                          bt.algos.SelectMomentum(n=6),
                          OrderedWeights([w,w,w,w,w,w]),
                          bt.algos.Rebalance()])
    
    return bt.Backtest(s,data)

tickers = 'BKLN, PRCIX, VFIIX, LUBIX, PRHYX, PRFHX, VUSTX, RPIBX, FCVSX, PSHAX, FNMIX, ECGOX, VBLTX, TLT, JNK, HYG, MUB, IEF'

meanVar = meanVar(tickers, mons = 3, name='Bond MeanVar', x=0.0, y=1.0)
nSix = nSix(tickers,name='nSix', w =.166)
nThree = nThree(tickers, name='nThree', w = .33)
benchmark = long_only_ew('TLT', name = 'TLT')
benchmark2 = long_only_ew('SPY', name = 'SPY')

res2 = bt.run(meanVar, benchmark, benchmark2)

print("Done")
```
